```python
df = pd.get_dummies(df)
```

```python
#Separando variaveis independentes
X=df.drop('SalePrice',axis=1)
y=df['SalePrice']
```

```python
#Criei o traint test para separar nossa base em treino e teste de forma aleatorioa
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Vamos padronizar nossas variaves continuas com ColumnTransformer do Sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

colunas_variaves_continuas = ['MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual', 'OverallCond',
       'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2',
       'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
       'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
       'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF',
       'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal',
       'MoSold', 'YrSold']


def setSacaler():
  ct = ColumnTransformer([('padronizar', StandardScaler(), colunas_variaves_continuas)], remainder='passthrough')
  return ct

scaler = setSacaler()
```

```python
scaler.fit(X_train)
```

```python
pd.DataFrame(scaler.transform(X_train), columns=X_train.columns).describe().T
```
